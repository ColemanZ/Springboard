**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#creating a function to compare prices from 1998 to 2018. Not sure why this random 10 year span is a good indicator of total picture..
#converted average price to float for ratio portion
def ratio_1998vs2018(B):
    y1998 = float(B['avg_price'][B['Year']==1998])
    y2018 = float(B['avg_price'][B['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
#empty dictionary
mean_dict = {}
```

```python
#loop to use my function on all the buroughs
#looping through the list of unique burough names rather than all the instances
for L in dfg['London_Borough_Name'].unique():
    #adding to my empty dictionary for each loop
    mean_dict[L] = ratio_1998vs2018(dfg[dfg['London_Borough_Name']==L])
print (mean_dict)
```

```python
#dataframe containing information above
ratios_df = pd.DataFrame(mean_dict)
ratios_df.head()
```

```python
#transposing for easier view
ratios_DF = ratios_df.T.reset_index()
```

```python
ratios_DF.head()
```

```python
#adding column names
ratios_DF.rename(columns={'index':'Borough',0:'1998vs2018_Ratio'}, inplace = True)
ratios_DF.head()
```

```python
#Sorting by fastest 50%
tophalf = ratios_DF.sort_values(by='1998vs2018_Ratio',ascending = False).head(16)
print(tophalf)
```

```python
#creating plot with top half 
tophalfplt = tophalf[['Borough','1998vs2018_Ratio']].plot(kind='bar')
#making bottom of chart readable by flipping the tick labels on the x axis
tophalfplt.set_xticklabels(tophalf.Borough)
```

```python
#bottom half for fun
bottomhalf = ratios_DF.sort_values(by='1998vs2018_Ratio',ascending = True).head(16)
bottomhalfplt = bottomhalf[['Borough','1998vs2018_Ratio']].plot(kind='bar')
bottomhalfplt.set_xticklabels(bottomhalf.Borough)
```
